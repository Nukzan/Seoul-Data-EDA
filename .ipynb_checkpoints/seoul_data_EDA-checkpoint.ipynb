```python
# 지하철 개수 데이터

public_mtr_ttl.head(5)
```

```python
# 2019년 공공자전거 데이터

public_bcl_new = public_bcl_ttl[public_bcl_ttl['기간'] == 2019]
public_bcl_new = public_bcl_new.drop('기간', axis=1)
public_bcl_new.head(5)
```

```python
# 2019년 보행만족도 데이터

crs_env_new = crs_env[crs_env['기간'] == 2019]
crs_env_new = crs_env_new[['자치구','종합']]
crs_env_new.head(5)
```

```python
# 2019년 주차장확보율 데이터

car_prk_new = car_prk_rt_region[car_prk_rt_region['기간'] == 2019]
car_prk_new = car_prk_new[['자치구','주차장확보율']]
car_prk_new.head(5)
```

```python
# 2019년 자치구별 인구 데이터

popul_new = population[population['기간'] == 2019]
popul_new = popul_new[['자치구','계']]
popul_new.head(5)
```

```python
# 데이터 병합

ttl = pd.merge(public_bus_ttl, public_mtr_ttl, how='outer', on='자치구')
ttl = pd.merge(ttl, public_bcl_new, how='outer', on='자치구')
ttl = pd.merge(ttl, crs_env_new, how='outer', on='자치구')
ttl = pd.merge(ttl, car_prk_new, how='outer', on='자치구')
ttl = pd.merge(ttl, popul_new, how='outer', on='자치구')
ttl
```

```python
# df 컬럼 이름 변경

ttl.rename(columns={ttl.columns[1]:'버스정거장개수',ttl.columns[2]:'지하철개수',ttl.columns[3]:'공공자전거거치대수',ttl.columns[4]:'보행만족도',ttl.columns[6]:'인구수',}, inplace=True)
ttl
```

## 인구수로 데이터 나누기

- 자치구별 비교를 위해서

```python
ttl['버스정거장'] = ttl['버스정거장개수'] / ttl['인구수']
ttl['지하철'] = ttl['지하철개수'] / ttl['인구수']
ttl['공공자전거'] = ttl['공공자전거거치대수'] / ttl['인구수']
ttl['보행안전'] = ttl['보행만족도'] / ttl['인구수']
ttl['주차시설'] = ttl['주차장확보율'] / ttl['인구수']
ttl.head(10)
```

```python
ttl_pop = ttl[['자치구','버스정거장','지하철','공공자전거','보행안전','주차시설']]
ttl_pop
```

## 스케일링

- minmax scaler 사용한 전체 데이터 스케일링

```python
ttl.columns
```

```python
# 자치구와 인구수 컬럼을 제외한 나머지 컬럼의 스케일링

score = pd.DataFrame(minmax_scale(ttl.iloc[:,7:], axis=0), columns=['버스정거장_s', '지하철_s', '공공자전거_s', '보행안전_s', '주차시설_s'])
score
```

```python
final = pd.concat([ttl_pop, score], axis=1)
final
```

```python
fig, axes = plt.subplots(3,2, figsize=(20,10))
plt.subplot(321, title='버스정류소')
plt.hist(final['버스정거장'])
plt.subplot(322, title='지하철')
plt.hist(final['지하철'])
plt.subplot(323, title='공공자전거')
plt.hist(final['공공자전거'])
plt.subplot(324, title='보행안전만족도')
plt.hist(final['보행안전'])
plt.subplot(325, title='주차장')
plt.hist(final['주차시설'])
plt.show()
```
