### 버스 정류소

```python
import numpy as np

# x값과 y값
x = [2,4,6,8]
y = [81,93,91,97]

# x와 y의 평균값
mx = np.mean(x)
my = np.mean(y)

# 기울기 공신의 분모와 분자
divisor = sum([(mx - i) **2 for i in x])

def top(x, mx, y, my):
    d = 0
    for i in range(len(x)):
        d += (x[i] - mx) * (y[i] - my)
    return d
dividend = top(x, mx, y, my)

# 기울기와 y절편 구하기
a = dividend / divisor
b = my - (mx*a)

# 확인
print('기울기= ', a)
print('y절편 = ', b)
```
