## 데이터 취합

- 년도별로 흩어진 데이터 취합 (보행만족도 조사 2010~2019)

```
# 년도별/자치구별 데이터만 추출

c2010 = crs_env_2010.loc[crs_env_2010[crs_env_2010['분류']=='종로구'].index.tolist()[0]:]
c2011 = crs_env_2011.loc[crs_env_2011[crs_env_2011['분류']=='종로구'].index.tolist()[0]:]
c2012 = crs_env_2012.loc[crs_env_2012[crs_env_2012['분류']=='종로구'].index.tolist()[0]:]
c2013 = crs_env_2013.loc[crs_env_2013[crs_env_2013['분류']=='종로구'].index.tolist()[0]:]
c2014 = crs_env_2014.loc[crs_env_2014[crs_env_2014['분류']=='종로구'].index.tolist()[0]:]
c2015 = crs_env_2015.loc[crs_env_2015[crs_env_2015['분류']=='종로구'].index.tolist()[0]:]
c2016 = crs_env_2016.loc[crs_env_2016[crs_env_2016['분류']=='종로구'].index.tolist()[0]:]
c2017 = crs_env_2017.loc[crs_env_2017[crs_env_2017['분류']=='종로구'].index.tolist()[0]:]
c2018 = crs_env_2018.loc[crs_env_2018[crs_env_2018['분류']=='종로구'].index.tolist()[0]:]
c2019 = crs_env_2019.loc[crs_env_2019[crs_env_2019['분류']=='종로구'].index.tolist()[0]:]
```

```
# 보행환경 만족도 연별/자치구별 데이터 취합
crs_env = pd.concat([c2010, c2011, c2012, c2013, c2014,
                     c2015, c2016, c2017, c2018, c2019])
```
